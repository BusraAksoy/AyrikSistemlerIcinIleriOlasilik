**Örnek:** Bir çekilişte toplam 10 bilet satılacak ve bu biletlerden 4'üne birer hediye çıkacaktır. Her bilete hediye çıkma olasılığı eşittir. 

**Buna göre, bir kişinin aldığı 3 bilete 2 hediye çıkma olasılığı kaçtır?**



Hediyesiz bilet : 6 \
hediyeli bilet : 4 \

seçilecek olan 3 biletten 2 bilete hediye çıkma durumu:

C( 4 , 2 ) = 4! / ( 4 - 2 )!.2! = 6 


Toplam 3 bilet alınacak. 2 si hediyelilerden alındı. Kalan 1 tanesi de hediyesiz biletlerden alınacak

Hediyesiz bilet : 6

C( 6 , 1 ) = 6! / ( 6 - 1 )!.1! = 6





Toplamda 10 bilet var, bunlardan 3 tanesi seçilecek.

C(10 , 3) = 10! / (10-3)!.3! = 120


olasılık: 

P ( A ) = İstenilen durumların sayısı / Tüm durumların sayısı

C (4, 2) . C ( 6, 1 ) / C(10 , 3) = 6.6 / 120 = 0,3





```python
def faktoriyel(f):
  if f<0:
    return "negatif sayıların faktoriyeli hesaplanamaz"
  elif f==1 or f==0:
    return 1
  else:
    return f*faktoriyel(f-1)

def kombinasyon(n,r):
  k=0
  if r>n:
    print("Sıralama sayısı(r) eleman sayısından (n) büyük olamaz")
  else:
    k=faktoriyel(n)/(faktoriyel(n-r)*faktoriyel(r))
  return k

# olasılık
# istenen/tüm durum

k1=kombinasyon(4,2)
k2=kombinasyon(6,1)
k3=kombinasyon(10,3)

olasilik= k1*k2/k3

print(olasilik)
```
