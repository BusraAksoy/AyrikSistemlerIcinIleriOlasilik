PERMÜTASYON

n ve r birer doğal sayı ve n ≥ r olmak üzere n elemanlı bir kümenin birbirinden farklı r tane elemanından oluşan dizilişlerin her birine n’nin r’li permütasyonu (dizilişi) denir.

permutasyon.png



**Örnek:** 5 arkadaş bir sıraya ikişerli oturup fotoğraf çektirecektir. Fotoğraf çekimi kaç farklı şekilde yapılabilir ?

5 kişinin ikişerli dizilişlerinin (permütasyonlarının) sayısı P (5, 2) ile bulunur.

P(5,2) = 5! / (5−2)! = 5! / 3! = 120 / 6 = 20

```python
def faktoriyel(f):
  if f<0:
    return "negatif sayıların faktoriyeli hesaplanamaz"
  elif f==1 or f==0:
    return 1
  else:
    return f*faktoriyel(f-1)

def permutasyon(n,r):
  p=0
  if r>n:
    print("Sıralama sayısı(r) eleman sayısından (n) büyük olamaz")
  else:
    p=faktoriyel(n)/faktoriyel(n-r)
  return p

print(permutasyon(5,2))
```

**Örnek:** A = { 1, 2, 3, 4, 5, 6 } kümesinin iki elemanlı alt kümelerinin (2’li kombinasyonlarının) sayısını bulun.

C(6,2) = 6! / (6−2)!.2! = 6! / 4!.2! = 720 / 48 = 15

```python
def faktoriyel(f):
  if f<0:
    return "negatif sayıların faktoriyeli hesaplanamaz"
  elif f==1 or f==0:
    return 1
  else:
    return f*faktoriyel(f-1)

def kombinasyon(n,r):
  k=0
  if r>n:
    print("Sıralama sayısı(r) eleman sayısından (n) büyük olamaz")
  else:
    k=faktoriyel(n)/(faktoriyel(n-r)*faktoriyel(r))
  return k

print(kombinasyon(6,2))
```
