Bu bilgilere gÃ¶re test sonucu pozitif olan bir Ã§ocuÄun gerÃ§ekten hasta olma olasÄ±lÄ±ÄÄ± nedir?


P(H) :  1/43 = 0,023 --->  hastalÄ±k olasÄ±ÄÄ±lÄ±

P(S) :  42/43 = 0,976 --->  hasta olmama olasÄ±lÄ±ÄÄ±. Yani saÄlÄ±klÄ± olmadurumu

P(p|H) : %80 = 0,8 --->  hastalÄ±klÄ± bir Ã§ocuÄun testinin pozitif olmasÄ±

P(p|S) : %10 = 0,1 --->  saÄlÄ±klÄ± bir Ã§ocugun testinin pozitif olmasÄ±

P(H|p) : ? --->  test sonucu pozitif olan bir Ã§ocugun hasta olma olasÄ±lÄ±ÄÄ±?



```
P(H|p) = P(p|H).P(H) / P(p) 

       = P(p|H).P(H) / (P(p|H).P(H) +P(p|S).P(S))

       = 0,8*0,023 / (0,8*0,023 + 0,1*0,976)

       = 0,1586
```




````python
def bayes(ph,ps):
  h=1/43
  s = 1-h
  hp = ph*h / (ph*h + ps*s)
  return hp

bayes(80/100,10/100)
````
